```python
import rawgpy
import requests
import pandas as pd
import numpy as np
import json
import csv
import os
import rawg
```

```python
game_df = pd.read_csv('../../DSBA6276_Customer_Analytics/Video_Game_Project/Video_Games_Sales_2013-2018_Full.csv')
game_df.head()
```

```python
def my_game_name_convert(x):
    if x['Name'] == 'CharadeManiacs':
        return 'Charade Maniacs'
    else:
        return x
```

```python
game_data = game_df.loc[(game_df.Year_of_Release >= 2013) | (game_df.Year_of_Release <= 2019)]
game_data_copy = game_data.copy()
```

```python
game_data_copy['Name'] = game_data_copy['Name'].replace('CharadeManiacs','Charade Maniacs').replace('WORKxWORK','WORK')
game_data_copy[game_data_copy['Name']=='Charade Maniacs']
game_data_copy[game_data_copy['Name']=='WORK']
```

```python
# Control Panel - Environment Variables
email    = os.environ.get('my_email')
password = os.environ.get('my_pass')
key_api  = os.environ.get('api_key')
```

```python
# api key and login credentials
rawg = rawgpy.RAWG(key_api) # your api key 
rawg.login(email, password) # your email and password 
```

```python

game_name                    = []
game_released                = []
game_rating                  = []
game_genres                  = []
game_playtime                = []

ratings_title                = []
ratings_count                = []
ratings_percent              = []

user_status_yet              = []
user_status_owned            = []
user_status_beaten           = []
user_status_toplay           = []
user_status_dropped          = []
user_status_playing          = []
user_status_metacritic       = []

user_status_score_yet        = []
user_status_score_owned      = []
user_status_score_beaten     = []
user_status_score_toplay     = []
user_status_score_dropped    = []
user_status_score_playing    = []
user_status_score_metacritic = []



for i in range(10):
    game              = rawg.get_request(f"https://api.rawg.io/api/games?dates=2013-01-01,2019-12-31&page={i+1}&page_size=1")
    df                = pd.DataFrame(game)
    g                 = df['results'][0]
    gplaytime         = g['playtime']
    gname             = g['name']
    greleased         = g['released']
    grating           = g['rating']
    ggenres           = g['genres'][0]['name']
    grate_title       = g['ratings'][0]['title']
    grate_count       = g['ratings'][0]['count']
    grate_percent     = g['ratings'][0]['percent']
    
    gstatus_yet       = list(g['added_by_status'].keys())[0]
    gstatus_owned     = list(g['added_by_status'].keys())[1]
    gstatus_beaten    = list(g['added_by_status'].keys())[2]
    gstatus_toplay    = list(g['added_by_status'].keys())[3]
    gstatus_dropped   = list(g['added_by_status'].keys())[4]
    gstatus_playing   = list(g['added_by_status'].keys())[5]
    
    gstatus_score_yet     = g['added_by_status']['yet']
    gstatus_score_owned   = g['added_by_status']['owned']
    gstatus_score_beaten  = g['added_by_status']['beaten']
    gstatus_score_toplay  = g['added_by_status']['toplay']
    gstatus_score_dropped = g['added_by_status']['dropped']
    gstatus_score_playing = g['added_by_status']['playing']

    
    user_status_yet.append(gstatus_yet)
    user_status_owned.append(gstatus_owned)
    user_status_beaten.append(gstatus_beaten)
    user_status_toplay.append(gstatus_toplay)
    user_status_dropped.append(gstatus_dropped)
    user_status_playing.append(gstatus_playing)
    
    user_status_score_yet.append(gstatus_score_yet)
    user_status_score_owned.append(gstatus_score_owned)
    user_status_score_beaten.append(gstatus_score_beaten)
    user_status_score_toplay.append(gstatus_score_toplay)
    user_status_score_dropped.append(gstatus_score_dropped)
    user_status_score_playing.append(gstatus_score_playing)

    
    game_name.append(gname)
    game_released.append(greleased)
    game_rating.append(grating)
    game_genres.append(ggenres)
    game_playtime.append(gplaytime)
    ratings_title.append(grate_title)
    ratings_count.append(grate_count)
    ratings_percent.append(grate_percent)


```

```python
game_name_df           = pd.DataFrame(game_name, columns=['Game Names'])
game_released_df       = pd.DataFrame(game_released,columns=['Released Date'])
game_rating_df         = pd.DataFrame(game_rating, columns=['Game Rating'])
game_genre_df          = pd.DataFrame(game_genres,columns=['Game Genres'])
game_playtime_df       = pd.DataFrame(game_playtime,columns=['playtime'])
game_rating_title_df   = pd.DataFrame(ratings_title,columns=['Title Ratings'])
game_rating_count_df   = pd.DataFrame(ratings_count,columns=['Ratings Count'])
game_rating_percent_df = pd.DataFrame(ratings_percent,columns=['Ratings Percents'])

game_status_yet        = pd.DataFrame(user_status_yet,columns=['User Status Yet'])
game_status_owned      = pd.DataFrame(user_status_owned,columns=['User Status Owned'])
game_status_beaten     = pd.DataFrame(user_status_beaten,columns=['User Status Beaten'])
game_status_toplay     = pd.DataFrame(user_status_toplay,columns=['User Status Toplay'])
game_status_dropped    = pd.DataFrame(user_status_dropped,columns=['User Status Dropped'])
game_status_playing    = pd.DataFrame(user_status_playing,columns=['User Status Playing'])

game_score_yet         = pd.DataFrame(user_status_score_yet,columns=['User Score Yet'])
game_score_owned       = pd.DataFrame(user_status_score_owned,columns=['User Score Owned'])
game_score_beaten      = pd.DataFrame(user_status_score_beaten,columns=['User Score Beaten'])
game_score_toplay      = pd.DataFrame(user_status_score_toplay,columns=['User Score Toplay'])
game_score_dropped     = pd.DataFrame(user_status_score_yet,columns=['User Score Dropped'])
game_score_playing     = pd.DataFrame(user_status_score_playing,columns=['User Score Playing'])


df_list = [ game_name_df
           ,game_genre_df
           ,game_released_df
           ,game_playtime_df
           ,game_rating_df
           ,game_rating_title_df
           ,game_rating_count_df
           ,game_rating_percent_df
           ,game_status_yet
           ,game_score_yet
           ,game_status_owned
           ,game_score_owned
           ,game_status_beaten
           ,game_score_beaten
           ,game_status_toplay
           ,game_score_toplay
           ,game_status_dropped
           ,game_score_dropped
           ,game_status_playing
           ,game_score_playing]

video_games = pd.concat(df_list,axis=1)
video_games
```

```python
yet = video_games.drop(['User Status Owned'
                        ,'User Score Owned'
                        ,'User Status Beaten'
                        ,'User Score Beaten'
                        ,'User Status Toplay'
                        ,'User Score Toplay'
                        ,'User Status Dropped'
                        ,'User Score Dropped'
                        ,'User Status Playing'
                        ,'User Score Playing'],axis=1).rename(columns={'User Status Yet': "User Status"
                                                                     ,'User Score Yet': "User Score"})

Owned = video_games.drop(['User Status Yet'
                        ,'User Score Yet'
                        ,'User Status Beaten'
                        ,'User Score Beaten'
                        ,'User Status Toplay'
                        ,'User Score Toplay'
                        ,'User Status Dropped'
                        ,'User Score Dropped'
                        ,'User Status Playing'
                        ,'User Score Playing'],axis=1).rename(columns={'User Status Owned': "User Status"
                                                                     , 'User Score Owned': "User Score"})


Beaten = video_games.drop(['User Status Yet'
                        ,'User Score Yet'
                        ,'User Status Owned'
                        ,'User Score Owned'
                        ,'User Status Toplay'
                        ,'User Score Toplay'
                        ,'User Status Dropped'
                        ,'User Score Dropped'
                        ,'User Status Playing'
                        ,'User Score Playing'],axis=1).rename(columns={'User Status Beaten': "User Status"
                                                                     , 'User Score Beaten': "User Score"})

Toplay = video_games.drop(['User Status Yet'
                        ,'User Score Yet'
                        ,'User Status Owned'
                        ,'User Score Owned'
                        ,'User Status Beaten'
                        ,'User Score Beaten'
                        ,'User Status Dropped'
                        ,'User Score Dropped'
                        ,'User Status Playing'
                        ,'User Score Playing'],axis=1).rename(columns={'User Status Toplay': "User Status"
                                                                     , 'User Score Toplay': "User Score"})

Dropped = video_games.drop(['User Status Yet'
                        ,'User Score Yet'
                        ,'User Status Owned'
                        ,'User Score Owned'   
                        ,'User Status Beaten'
                        ,'User Score Beaten'
                        ,'User Status Toplay'
                        ,'User Score Toplay'
                        ,'User Status Playing'
                        ,'User Score Playing'],axis=1).rename(columns={'User Status Dropped': "User Status"
                                                                     , 'User Score Dropped': "User Score"})


Playing = video_games.drop(['User Status Yet'
                        ,'User Score Yet'
                        ,'User Status Owned'
                        ,'User Score Owned'
                        ,'User Status Beaten'
                        ,'User Score Beaten'
                        ,'User Status Toplay'
                        ,'User Score Toplay'
                        ,'User Status Dropped'
                        ,'User Score Dropped'],axis=1).rename(columns={'User Status Playing': "User Status"
                                                                     , 'User Score Playing': "User Score"})
```

```python
df_list_long = [yet
               ,Owned
               ,Beaten
               ,Toplay
               ,Dropped
               ,Playing]

video_games_long = pd.concat(df_list_long,axis=0).reset_index().drop('index',axis =1)
video_games_long.head()
```

## All Video Games In Original Dataset 

```python
release_date   = []
game_names     = []
playtimes      = []
ratings        = []
reviews_counts = []
metacritics    = []
for i in game_data_copy['Name'].unique()[:len(game_data_copy['Name'].unique())]:
    games             = rawg.get_request(f"https://api.rawg.io/api/games?search={i}&page=1&page_size=1")
   
    gplaytime         = games['results'][0]['playtime']
    gname             = games['results'][0]['name']
    greleased         = games['results'][0]['released']
    grating           = games['results'][0]['rating']
    reviews_count     = games['results'][0]['reviews_count']
    metacritic        = games['results'][0]['metacritic']
    
    
    release_date.append(greleased)
    game_names.append(gname)
    playtimes.append(gplaytime )
    ratings.append(grating)
    metacritics.append(metacritic)
    reviews_counts.append(reviews_count)
```

```python
game_name_df      = pd.DataFrame(game_names, columns=['Game Names'])
game_released_df  = pd.DataFrame(release_date,columns=['Released Date'])
game_rating_df    = pd.DataFrame(ratings, columns=['Game Rating'])
game_plytime_df   = pd.DataFrame(playtimes,columns=['Playtimes'])
reviews_counts_df = pd.DataFrame(reviews_counts, columns=['Review Counts'])
metacritics_df    = pd.DataFrame(metacritics,columns=['Metacritics Score'])
```

```python
df_list = [ game_name_df
           ,game_released_df
           ,game_rating_df
           ,reviews_counts_df
           ]

video_games = pd.concat(df_list,axis=1)
video_games
```
